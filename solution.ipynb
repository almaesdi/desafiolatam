## Generacion de caracteristicas adicionales

```python
#Definicion de funciones:

def isTemporadaAlta(fecha):

    #Obtengo el año de la fecha
    year = int(fecha.year)

    if fecha >= date(year=year, month=12, day=15) and fecha <= date(year=(year+1), month=3, day=3):
        return 1
    if fecha >= date(year=(year-1), month=12, day=15) and fecha <= date(year=year, month=3, day=3):
        return 1
    if fecha >= date(year=year, month=7, day=15) and fecha <= date(year=year, month=7, day=31):
        return 1
    if fecha >= date(year=year, month=9, day=11) and fecha <= date(year=year, month=9, day=30):
        return 1
    return 0

def getPeriodoDia(fecha):
    hora = fecha.time()

    if hora >= time(5,00) and hora < time(12,00):
        return 'mañana'
    if hora >= time(12,00) and hora < time(19,00):
        return 'tarde'
    #if (hora >= time(19,00) and hora <= time(00,00)) or (hora >= time(00,00) and hora <= time(4,59)):
    if hora >= time(19,00) or hora <= time(4,59):
        return 'noche'
    
    return "undefined"
```

Calcularemos el tiempo delta entre la hora programada y la de operacion

```python
df["diff_min"] = (df['Fecha-O'] - df["Fecha-I"]).dt.total_seconds() / 60
```

```python
df["diff_min"].describe()
```

¿Por qué hay valores negativos?¿-14 minutos?

```python
df[df["diff_min"] < 0]
```

**Se asume que existen vuelos que salen antes de lo programado, pero no suponen un problema**

```python
df["temporada_alta"] = df['Fecha-I'].apply(lambda s: isTemporadaAlta(s.date()))
```

```python
df["periodo_dia"] = df['Fecha-I'].apply(lambda s: getPeriodoDia(s))
```

```python
df
```

```python
df['atraso_15'] = np.where(df['diff_min'] > 15, 1, 0)
```

```python
#Verifico las transformaciones
print(df['temporada_alta'].isna().sum())
print(df['temporada_alta'].unique())

print(df['periodo_dia'].isna().sum())
print(df['periodo_dia'].unique())

print(df['atraso_15'].isna().sum())
print(df['atraso_15'].unique())
```

```python
#Guardamos las variables adicionales
df[['diff_min','temporada_alta','periodo_dia','atraso_15']].to_csv("synthetic_features.csv")
```
